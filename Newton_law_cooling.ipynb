# Newton's Law of Cooling

```python

#Initial conditions
dt = 1 #minutes
period = 121 #minutes
T_cup_pred = [68] # degrees celsius
t_cup_pred = range(period) #time in minutes
T_amb = 25.4 # degrees celsius
K = 0.027 

```

```python
# Implement the model
for t in range(1,period):
    dT = -K * (T_cup_pred[t-1]- T_amb) * dt
    T_cup_pred_updated = T_cup_pred[t-1] + dT
    T_cup_pred.append(T_cup_pred_updated)
    

```

```python
t_cup_obs = [0,3,6,9,12,18,24,30]
T_cup_obs = [68,64,60.3,56.8,54.5,49.6,46.8,44.1]
```

```python
import matplotlib.pyplot as plt

plt.figure( figsize=(8,5) )
plt.plot(t_cup_pred, T_cup_pred, "--r")
plt.plot(t_cup_obs, T_cup_obs, ' ob')
plt.xlabel('Time (minutes)', size=16)
plt.ylabel('Temeprature (Celicous)', size=16)

plt.show()
```
