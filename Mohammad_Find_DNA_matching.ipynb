```python
# load file, and handle extra header lines in file
dna = open('data\dna_sequence.txt').read().split('\n')
dna = dna[8:]

# make a single long string 
singleLong = [lt for wd in dna for lt in wd]
```

```python
# define the dnamaching function
Complementary  = ''
def dnamatching (dna):
    Complementary  = ''    
    for value in singleLong:
        if value == "G":
            Complementary  = Complementary  + 'T'
        elif value == "T":
            Complementary  = Complementary  + 'G'
        elif value == "C":
            Complementary  = Complementary  + 'A'
        elif value == "A":
            Complementary  = Complementary  + 'C'
        else:
            print ('Something went wrong, unknown base found:', value)
            break
        
    return Complementary 
```
