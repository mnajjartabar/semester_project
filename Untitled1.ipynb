```python
import pandas as pd

url1 = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv'
url2 = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv'
url3 = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv'
infected = pd.read_csv(url1,index_col=0)
death = pd.read_csv(url2,index_col=0)
recovered =pd.read_csv(url3,index_col=0)

infected.head()
```

```python
sum_infected = infected.sum(axis = 0,skipna = True)
sum_death = death.sum(axis = 0, skipna = True)
sum_recovered = recovered.sum(axis = 0, skipna = True)


#sum_infected.drop("Province/State")
sum_infected
```

```python
sum_infected=sum_infected.drop("Lat")
sum_infected=sum_infected.drop("Long")
sum_infected=sum_infected.drop("Country/Region")

sum_death=sum_death.drop("Lat")
sum_death=sum_death.drop("Long")
sum_death=sum_death.drop("Country/Region")


sum_recovered=sum_recovered.drop("Lat")
sum_recovered=sum_recovered.drop("Long")
sum_recovered=sum_recovered.drop("Country/Region")


```

```python
df = pd.DataFrame([sum_infected, sum_death, sum_recovered])
df
```

```python
df = df.T
```
